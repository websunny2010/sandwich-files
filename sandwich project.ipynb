```python
available_sandwiches=['chicken','aaloo tikki','paneer','veg loaded']
available_fillings=['mushroom','cheese','olives','jalapeno','pickle']
sandwich_prices={'chicken':20,'aaloo tikki':10,'paneer':15,'veg loaded':12}
fillings_prices={'mushroom':10,'cheese':5,'olives':7,'jalapeno':5,'pickle':4}

subtotal=[]
final_order={}
customer_address={}

print('Hi! Welcome to our text based sandwich ordering system.')

order_sandwich = True

while order_sandwich:
    print('Please choose a sandwich: ')
    print()
 
    for sandwiches in available_sandwiches:
        print()
        print(sandwiches)
        print()
    while True:
 
       sandwich = input ('\nWhich sandwich would you like to order: ')
 
       if sandwich in available_sandwiches:
         print(f'\nYou have ordered {sandwich} sandwich')
 
         final_order.setdefault(sandwich,[])
 
         break
 
       if sandwich not in available_sandwiches:
            print()
            print(f'\nSorry we do not have {sandwich} sandwich at the moment.')
            print()
 
 
    print('This is the list of available fillings: ')
    print()
    for fillings in available_fillings:
        print()
        print(fillings)
 
 
# order_fillings = True
 
    while True:
 
       extra_fillings=input('Would you like extra fillings? Yes or No: ')
    
       if extra_fillings=='Yes' or extra_fillings=='yes':
            
            filling = input('Which filling would you like to add?: ')
 
            if filling in available_fillings:
               print(f'We have added {filling} filling.')
 
               final_order[sandwich].append(filling)
 
            else:
                print(f'Sorry we do not have {filling} filling at the moment.')
 
       elif extra_fillings=='No' or extra_fillings=='no':
 
           break
 
    extra_sandwich=input("Would you like to order an extra sandwich?")
 
    if extra_sandwich=='No' or extra_sandwich=='no':
 
       for key,value in final_order.items():
 
           print(f'\nYou have ordered {key} sandwich with {value} filling.')
 
       check_order = True
 
       while check_order:
          print()
    
          order_correct = input('Is this correct? yes or no')
 
          if order_correct=='Yes' or order_correct=='yes':
 
             check_order=False
             order_sandwich=False
 
          if order_correct=='No' or order_correct=='no':
 
              print(final_order)
 
              add_remove = input ('Would you like to add or remove?')
 
              if add_remove=='remove' or add_remove=='Remove':
 
                  remove=input('Which sandwich would you like to remove?')
 
                  del final_order[remove]
 
                  print(final_order)
              if add_remove=='Add' or add_remove=='add':
                  check_order=False
print()

print(final_order)
sandwich_order=[]
filling_order=[]
for i in final_order.keys():
 #print(i,end=', ')
 sandwich_order.append(i) 
print()
for k in final_order.keys():
    for j in final_order.get(k):
 #print(j,end=', ')
     filling_order.append(j)
for sw in sandwich_order:
    subtotal.append(sandwich_prices.get(sw))
    for fil in filling_order:
        subtotal.append(fillings_prices.get(fil))
print(f'\nYour total order price is Rs. {sum(subtotal)}')
print("Please provide us with your name, address and phone number.")
customer_address['name']=input('Your name please: ')
customer_address['street_name']=input('Street name and house number: ')
customer_address['postalcode']=input('Postal code and city name: ')
customer_address['phonenumber']=input('Your number: ')
print()
print(f"Thank you for your order {customer_address['name']}.")
print()
print(f'We wil deiver your order to this address.')
print()
print(customer_address['street_name'])
print(customer_address['postalcode'])
print()
print(f"We will contact you on {customer_address['phonenumber']} if anything comes")
```
